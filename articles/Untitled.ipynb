# Pandemic Network Centrality

```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt

import random
from collections import defaultdict
```

```python
df = pd.read_csv("pandemic.csv", index_col=0)
```

```python
df.head()
```

```python
df.sum().value_counts().plot.bar()
```

## Random Walk

```python
connections = defaultdict(list)
for city in df.columns:
    col = df[city]
    dest = list(col[col > 0].index)
    connections[city] = dest
```

```python
city = "Atlanta"
freq = defaultdict(int)

for _ in range(10_000_000):
    city = random.choice(connections[city])
    freq[city] += 1
```

```python
s=pd.Series(freq)
s = s / s.sum()
s.sort_values(ascending=False, inplace=True)
s.head(10)
```

```python
s.plot.bar()
s.sum()
```

## Eigenvectors

```python
mtx = df.values
eig_vec, eig_mtx = np.linalg.eig(mtx)
eig_mtx.shape
```

```python
dom = eig_mtx[:,0]
dom = dom / dom.sum()
dom.sum()
```

```python
s2 = pd.Series(dom, index=df.columns)
s2.sort_values(ascending=False, inplace=True)
s2.plot.bar()
```
