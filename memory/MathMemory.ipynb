```python
s = pd.Series([12, 6, 4, 2], index=list("    "))
s.plot.pie(explode=[0, 0, 0, 0.2])
```

```python
plt.figure(figsize=(5, 5))
s.plot.bar()
plt.xlabel("")
plt.ylabel("")
plt.xticks([])
plt.yticks([])
```

```python
flights = sns.load_dataset("flights")
plt.figure(figsize=(5, 5))
flights["passengers"][36:68].plot(linewidth=7)
plt.xlabel("")
plt.ylabel("")
plt.xticks([])
plt.yticks([])
plt.savefig("timeseries.png")
```

```python
pingus = sns.load_dataset("penguins")
plt.figure(figsize=(5, 5))
sns.scatterplot(data=pingus, x="bill_length_mm", y="bill_depth_mm", hue="species", s=100, legend=False)
plt.xlabel("")
plt.ylabel("")
plt.xticks([])
plt.yticks([])
plt.savefig("scatter.png")
```

```python
g = np.random.normal(loc=0.0, scale=1.0, size=100_000)
s = pd.Series(g)
s = s[(s > -4) & (s < 4)]
plt.figure(figsize=(5, 5))
s.hist(bins=40)
plt.xlabel("")
plt.ylabel("")
plt.xticks([])
plt.yticks([])
plt.savefig("histo.png")
```

```python
plt.figure(figsize=(5, 5))
p = pingus.iloc[-125:].copy()
p.loc[219, "body_mass_g"] = 4400
p.loc[219, "bill_depth_mm"] = 16.5
sns.scatterplot(data=p, x="body_mass_g", y="bill_depth_mm", hue="species", s=100, legend=False)
plt.xlabel("")
plt.ylabel("")
plt.xticks([])
plt.yticks([])
plt.savefig("outlier.png")
```
