# Stats Memory: Plots

* pie chart
* bar plot
* box plot
* histogram
* scatter plot
* line plot
* heat map

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
pingus = sns.load_dataset("penguins")
```

```python
plt.figure(figsize=(5, 5))
s = pd.Series([12, 6, 4, 2], index=list("    "))
s.plot.pie(explode=[0, 0, 0, 0.2])
plt.savefig("piechart.png")
```

```python
plt.figure(figsize=(5, 5))

pingus.groupby("species")["sex"].count().plot.bar()
plt.xlabel("")
plt.ylabel("")
plt.xticks([])
plt.yticks([])
plt.savefig("barchart.png")
```

```python
flights = sns.load_dataset("flights")
plt.figure(figsize=(5, 5))
flights["passengers"][36:68].plot(linewidth=7)
plt.xlabel("")
plt.ylabel("")
plt.xticks([])
plt.yticks([])
plt.savefig("timeseries.png")
```

```python
plt.figure(figsize=(5, 5))
sample = pingus.sample(150)
sns.scatterplot(data=sample, x="bill_length_mm", y="bill_depth_mm", hue="species", s=200, legend=False)
plt.xlabel("")
plt.ylabel("")
plt.xticks([])
plt.yticks([])
plt.savefig("scatter.png")
```

```python
g = np.random.normal(loc=0.0, scale=1.0, size=10_000)
s = pd.Series(g)
s = s[(s > -4) & (s < 4)]
plt.figure(figsize=(5, 5))
s.hist(bins=20)
plt.xlabel("")
plt.ylabel("")
plt.xticks([])
plt.yticks([])
plt.savefig("histo.png")
```

```python
plt.figure(figsize=(5, 5))
p = pingus.iloc[-125:].copy()
p.loc[219, "body_mass_g"] = 4400
p.loc[219, "bill_depth_mm"] = 16.5
sns.scatterplot(data=p, x="body_mass_g", y="bill_depth_mm", hue="species", s=400, legend=False)
plt.xlabel("")
plt.ylabel("")
plt.xticks([])
plt.yticks([])
plt.savefig("outlier.png")
```

```python
num = pingus.iloc[:,2:].copy().dropna()
num ["sex"] = (num["sex"] == "Male").astype(int)
cm = num.corr().abs()
plt.figure(figsize=(5, 5))
sns.heatmap(cm, cbar=False, cmap="viridis")
plt.xlabel("")
plt.ylabel("")
plt.xticks([])
plt.yticks([])
plt.savefig("heatmap.png")
```

```python
plt.figure(figsize=(5, 5))
sns.boxplot(data=pingus, y="flipper_length_mm", hue="species", gap=0.1, showfliers=False, legend=False, linewidth=3)
plt.xlabel("")
plt.ylabel("")
plt.xticks([])
plt.yticks([])
plt.savefig("boxplot.png")
```

```python
plt.figure(figsize=(5, 5))
p = pingus.iloc[:100].copy()
p["sample"] = "a"
sample = p.sample(10)
p.loc[sample.index.values, "sample"] = "b"
p.sort_values(by="sample", inplace=True)
sns.scatterplot(data=p, x="body_mass_g", y="bill_depth_mm", hue="sample", s=400, legend=False)
plt.xlabel("")
plt.ylabel("")
plt.xticks([])
plt.yticks([])
plt.savefig("sample.png")
```
